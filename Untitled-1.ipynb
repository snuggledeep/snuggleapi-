```python
def getStockData(stockSymbol, interval, numMonths):

    # Get the stock data from the API
    for i in range(1, numMonths):
        CSV_URL = "https://www.alphavantage.co/query?function=TIME_SERIES_INTRADAY_EXTENDED&symbol=" + stockSymbol + "&interval=" + interval + "&slice=year1month"+str(i)+"&apikey=UHXXK2NEVHOR2LSL"

        with requests.Session() as s:
            download = s.get(CSV_URL)

            print(download.status_code)

            decoded_content = download.content.decode('utf-8')
            # print(decoded_content)

            cr = csv.reader(decoded_content.splitlines(), delimiter=',')
            my_list = list(cr)

            # Concatenate the dataframes
            if i == 1:
                df = pd.DataFrame(my_list[0:])
            else:
                df = pd.concat([df, pd.DataFrame(my_list[1:])])

    return df
```
